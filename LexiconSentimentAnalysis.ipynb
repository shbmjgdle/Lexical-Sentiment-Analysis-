```python
import io 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
#word_tokenize accepts a string as an input, not a file. 
stop_words = set(stopwords.words('english')) 
file1 = open('filter.csv','r',encoding='utf=8') 
line = file1.read()# Use this to read file content as a stream: 
words = line.split() 
for r in words: 
	if not r in stop_words: 
		appendFile = open('filteredtext.csv','a',encoding='utf=8') 
		appendFile.write(" "+r) 
		appendFile.close() 
```

```python
import csv
with open("filteredtext.csv", "rt",encoding='utf-8') as infile, open("out.csv", "wb",encoding='utf-8') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    conversion = set('_"/.$*@&^)(-!+/^[0-9]')
    for row in reader:
        newrow = [''.join('_' if c in conversion else c for c in entry) for entry in row]
        writer.writerow(newrow)
```
